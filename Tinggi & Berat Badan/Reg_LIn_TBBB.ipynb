# Import Library & Baca Data Serta Membuat DataFrame

```python
# Import library pandas untuk manipulasi dan analisis data
import pandas as pd
# Import library numpy untuk operasi matematik
import numpy as np
# Import LinearRegression dari sklearn untuk membuat model regresi linear
from sklearn.linear_model import LinearRegression
# Import matplotlib.pyplot dan seaborn untuk visualisasi data
import matplotlib.pyplot as plt
import seaborn as sns
#Mengecek Akurasi Data
from sklearn.metrics import r2_score
```

```python
# Data contoh berat badan (dalam kilogram) dan tinggi badan (dalam sentimeter)
data = {
    'Tinggi_Badan': [175, 160, 185, 170, 162, 165,169],
    'Berat_Badan': [70, 65, 80, 75, 68, 72, 77]
}

# Membuat DataFrame
dataframe = pd.DataFrame(data)
```

```python
dataframe
```

```python
plt.scatter(dataframe.Berat_Badan, dataframe.Tinggi_Badan, color='gray')
plt.xlabel('Tinggi Badan')
plt.ylabel('Berat Badan')
plt.title("Plot Data Tinggi dan Berat Badan")
plt.show()
```

# Korelasi

```python
dataframe.corr()
```

# Split

```python
tinggi_badan = dataframe['Tinggi_Badan'].values[:,np.newaxis]
berat_badan = dataframe['Berat_Badan'].values
```

# Modelling

```python
model_regresi = LinearRegression()

model_regresi.fit(tinggi_badan,berat_badan)
```

# Testing

```python
tb_test = [[20],[100]]
predik_bb = model_regresi.predict(tb_test)
```

```python
print("Prediksi Berat Badan : ", predik_bb)
```

```python
r2_bb = r2_score(berat_badan, model_regresi.predict(tinggi_badan))
print("R-squared Berat Badan", r2_bb)
```

# Visualisasi Data

```python
sns.pairplot(dataframe)
plt.suptitle("Pairplot Tinggi dan Berat Badan", y=1.02)
plt.show()
```

```python
# Visualisasi Hasil Prediksi Pengalaman dan Gaji
plt.scatter(dataframe.Tinggi_Badan, dataframe.Berat_Badan, color='gray', label='Data Asli')
plt.plot(tb_test, predik_bb, color='red', linewidth=2, label='Regresi Linear')
plt.xlabel('Tinggi Badan')
plt.ylabel('Berat Badan')
plt.title("Prediksi Tinggi Badan berdasarkan Berat Badan")
plt.legend()
plt.show()
```
